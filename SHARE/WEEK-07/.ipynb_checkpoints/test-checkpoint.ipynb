# Scatter plots 

In this example we show a basic interactive scatter plot generated using plotly 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

	plt.plot([1, 2, 3, 4],[1, 2, 3, 4],'o')

	#SET FIGURE LABELS 
	FS=18	 
	plt.xlabel('x', fontsize=FS)
	plt.ylabel('y', fontsize=FS)
	plt.show()
	#exit()
```
